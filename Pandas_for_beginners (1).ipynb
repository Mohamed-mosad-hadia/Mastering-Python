# Aggregations and Groupby

- Groupby() is a powerful function in pandas that allows you to group data based on a single column or more

- You can apply many operations to a groupby object, including aggregation functions like sum(), mean(), and count(), as well as lambda function and other custom functions using apply().

## groupby

1.  agg
2.  filter
3.  transform
4.  apply


````python
# Sample DataFrame (replace with your actual data)
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Value1': [10, 15, 20, 25, 30, 35],
        'Value2': [5, 10, 15, 20, 25, 30]}
df = pd.DataFrame(data)
df
````

````python
# 1. Aggregation (agg)
# aggfunc: can be  function, list of functions, dict, default numpy.mean
# Calculate the sum and mean of 'Value1' and 'Value2' for each category
aggregated_df = df.groupby('Category').agg({'Value1': ['sum', 'mean'], 'Value2': ['sum', 'mean']})
print("Aggregated DataFrame:\n", aggregated_df)
````

````python
# 2. Filtering (filter)
# Keep only categories where the mean of 'Value1' is greater than 20
filtered_df = df.groupby('Category').filter(lambda x: x['Value1'].mean() > 20)
print("\nFiltered DataFrame:\n", filtered_df)
````

````python
# 3. Transformation (transform)
# Standardize 'Value1' within each category (subtract mean, divide by standard deviation)
#Returns A DataFrame of the same shape as the input with transformed values.

df['Value1_standardized'] = df.groupby('Category')['Value1'].transform(lambda x: (x - x.mean()) / x.std())
print("\nDataFrame with Transformed Column:\n", df)
````

````python
# 4. Apply (apply)
# Create a new column 'Value_Ratio' by dividing 'Value1' by the sum of 'Value1' for each category.
def calculate_ratio(group):
    group['Value_Ratio'] = group['Value1'] / group['Value1'].sum()
    return group

applied_df = df.groupby('Category').apply(calculate_ratio)
print("\nDataFrame with Applied Function:\n", applied_df)
````

````python
# Example with multiple aggregate functions and custom function
def range_fn(x):
    return f"{x.min()}-{x.max()}"

grouped = df.groupby('Category').agg(
    value1_sum=('Value1','sum'),
    value1_range=('Value1',range_fn),
    value2_mean=('Value2','mean')
    )
print("\nDataFrame with multiple agg functions and custom function:\n",grouped)

# Example of using a custom function with apply
def custom_function(group):
    # Perform calculations on the group
    group['new_col'] = group['Value1'] * 2 + group['Value2']
    return group
result = df.groupby('Category').apply(custom_function)
print("\nDataFrame with custom function using apply:\n",result)
````

# Pivot Table

- A PivotTable is a powerful tool to calculate, summarize, and analyze data that lets you see comparisons, patterns, and trends in your data.

- pandas.pivot_table(

  data,

  values=None,

  index=None,

  columns=None,

  aggfunc=’mean’,

  fill_value=None,

  margins=False,

  dropna=True,

  margins_name=’All’

  )


````python
# prompt: pivot table explain

# Import pandas library
import pandas as pd

# Sample DataFrame (replace with your actual data)
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
        'Value1': [10, 15, 20, 25, 30, 35],
        'Value2': [5, 10, 15, 20, 25, 30]}
df = pd.DataFrame(data)

# Basic pivot table: Calculate the sum of 'Value1' for each 'Category' and 'Subcategory'
pivot_table = pd.pivot_table(df, values='Value1', index='Category', columns='Subcategory', aggfunc='sum')
print("Basic Pivot Table:\n", pivot_table)

# Pivot table with multiple aggregate functions: Calculate the sum and mean of 'Value1'
pivot_table_multiple_agg = pd.pivot_table(df, values='Value1', index='Category', columns='Subcategory', aggfunc=['sum', 'mean'])
print("\nPivot Table with Multiple Aggregate Functions:\n", pivot_table_multiple_agg)

# Pivot table with margins: Include row and column totals
pivot_table_margins = pd.pivot_table(df, values='Value1', index='Category', columns='Subcategory', aggfunc='sum', margins=True)
print("\nPivot Table with Margins:\n", pivot_table_margins)

# Pivot table with fill_value: Replace missing values with a specific value (e.g., 0)
pivot_table_fill_value = pd.pivot_table(df, values='Value1', index='Category', columns='Subcategory', aggfunc='sum', fill_value=0)
print("\nPivot Table with Fill Value:\n", pivot_table_fill_value)
````

# After cleaning your data, you can save it in different formats using **Pandas**. Here are the most common methods:

---

### **1️⃣ Save as CSV (Most Common)**

```python
df.to_csv("cleaned_data.csv", index=False)
```

✅ **Removes the index** (recommended for clean datasets).  
✅ **Fast and widely used** format.

If you **want to keep the index**, use:

```python
df.to_csv("cleaned_data.csv", index=True)
```

---

### **2️⃣ Save as Excel**

```python
df.to_excel("cleaned_data.xlsx", index=False)
```

✅ Best for sharing with **non-technical users**.  
✅ Supports **multiple sheets** (using `ExcelWriter`).

Example for multiple sheets:

```python
with pd.ExcelWriter("cleaned_data.xlsx") as writer:
    df.to_excel(writer, sheet_name="Sheet1", index=False)
```

---

### **3️⃣ Save as JSON**

```python
df.to_json("cleaned_data.json", orient="records")
```

✅ Useful for **web applications** and APIs.  
✅ Stores data as a list of dictionaries.

---

### **4️⃣ Save as Pickle (Fastest for Python)**

```python
df.to_pickle("cleaned_data.pkl")
```

✅ **Best for large datasets** (preserves DataFrame format).  
🚫 **Only works with Python** (not human-readable).

---

### **5️⃣ Save to a Database (SQL)**

```python
from sqlalchemy import create_engine

engine = create_engine("sqlite:///cleaned_data.db")
df.to_sql("table_name", engine, if_exists="replace", index=False)
```

✅ Best for **large datasets** and **long-term storage**.  
✅ Supports **querying with SQL**.

---

### **Which Format Should You Use?**

| **Format**  | **Best For**               |
| ----------- | -------------------------- |
| `.csv`      | General use & easy sharing |
| `.xlsx`     | Reports & spreadsheets     |
| `.json`     | Web apps & APIs            |
| `.pkl`      | Fast loading in Python     |
| `.db` (SQL) | Large datasets & databases |
